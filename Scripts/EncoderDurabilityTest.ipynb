```python
import cv2
from pathlib import Path
import os

OUT_DIR = '../outputs'
index = 0
while True:
    save =  'Trial - ' + str(index)
    outputPath = Path(f'{OUT_DIR}/{save}')
    if outputPath.is_dir():
        index += 1
        continue
    else:
        os.makedirs(f'{OUT_DIR}/{save}', exist_ok=False)
        OUT_DIR = OUT_DIR + f'/{save}'
        break

cap = cv2.VideoCapture('../Data/learnerUnderwater.mp4')
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
video_fps = int(cap.get(5)) 
if not cap.isOpened():
    print("Not opening")

out = cv2.VideoWriter(
    f'{OUT_DIR}/Video0.mkv', 
    cv2.CAP_FFMPEG,
    cv2.VideoWriter_fourcc(*"FFV1"), 
    video_fps, 
    (frame_width, frame_height), 
)
while cap.isOpened():
    ret, frame = cap.read()
    if ret:
        out.write(frame)
        # Press `q` to exit
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break
cap.release()
out.release()

for i in range(10):
    cap = cv2.VideoCapture(f'{OUT_DIR}/Video{i}.mkv')
    frame_width = int(cap.get(3))
    frame_height = int(cap.get(4))
    video_fps = int(cap.get(5))
    out = cv2.VideoWriter(
        f'{OUT_DIR}/Video{i + 1}.mkv', 
        cv2.CAP_FFMPEG,
        cv2.VideoWriter_fourcc(*"FFV1"),
        video_fps, 
        (frame_width, frame_height), 
    ) 
    if not cap.isOpened():
        print("Not opening")
    while cap.isOpened():
        ret, frame = cap.read()
        if ret:
            out.write(frame)
            # Press `q` to exit
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break
        else:
            break
    cap.release()
    out.release()

cap = cv2.VideoCapture(f'{OUT_DIR}/Video10.mkv')
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
video_fps = int(cap.get(5)) 
if not cap.isOpened():
    print("Not opening")

out = cv2.VideoWriter(
    f'{OUT_DIR}/FinalVid.mp4', 
    cv2.VideoWriter_fourcc(*"mp4v"), 
    video_fps, 
    (frame_width, frame_height), 
)
while cap.isOpened():
    ret, frame = cap.read()
    if ret:
        out.write(frame)
        # Press `q` to exit
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break
cap.release()
out.release()

```
